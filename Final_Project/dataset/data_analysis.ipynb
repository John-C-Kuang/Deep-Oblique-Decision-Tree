```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Setup DataFrame
df = pd.read_csv("Wine_Quality_Data.csv")
df
```

### Columns and Rows

```python
print("Rows: ", df.shape[0])
print("Columns: ", df.shape[1])
```

### Show the distribution of the target label (quality)

```python
quality = df["quality"]
quality

cnt = dict(Counter(quality))
items = list(cnt.items())

items = np.array(sorted(items, key=lambda item: item[0]))

keys = items[:, 0]
items = items[:, 1]

plt.bar(keys, items)
plt.xlabel("Wine Quality")
plt.ylabel("Instance count")

plt.show()
```

### Central Tendency

```python
qualities = []
for i in range(len(keys)):
    qualities.extend([keys[i]] * items[i])

mean = np.mean(qualities)
median = np.median(qualities)

mean, median
```
